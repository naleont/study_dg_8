## Цикл while

Представим себе задачу:\
Пользователь вводит слово из 5 букв, нужно напечатать это слово наоборот.\
Решение достаточно простое: нужно вызвать символы по их индексам от пятого к первому. Индексы можно перебирать с начала или с конца, это не имеет значения.

```python
word = input('Введите слово из 5 букв: ')

print(word[4] + word[3] + word[2] + word[1] + word[0])
```

или

```python
word = input('Введите слово из 5 букв: ')

print(word[-1] + word[-2] + word[-3] + word[-4] + word[-5])
```

А если букв не 5? Программа либо сломается, если букв меньше, либо напечатает слово не целиком, если оно длиннее.

Если писать программу для каждого варианта длины слова, это долго, занудно, и вообще теряется вся прелесть программирования.

Поэтому нам нужны циклы.

Циклы позволяют повторять выполнение одного и того же кода много раз: столько, сколько нам понадобится.

Цикл можно задать двумя командами, первая - `while`. Обратите внимание: скобок нет, значит это не функция, а оператор.

**Синтаксис.** После `while` пишется условие выполнения кода. Пока условие выполняется, код повторяется. Как только условие окажется ложным, программа выйдет из цикла. После условия - двоеточие, а с новой строки с отступом - код, который нужно будет выполнять повторно.

```python
a = 1

while a <= 5:
    print(a)
    a = a + 1
```

Обратите внимание на последнюю строку повторяющегося кода. Без изменения значения переменная всегда будет равна 1, а значит программа будет выполняться бесконечно. Не надо так.

А еще эту строку можно записать короче.

```python
a = 2
a += 3
print(a)
```

Так же можно записать и другие математические операции, в которых математическое действие производится с переменной.

```python
a = 3

a *= 3
print(a)

a -= 3
print(a)

a /= 3
print(a)
```

Вернемся к задаче. Пользователь вводит слово из 5 букв. Программа должна напечатать это слово наоборот.

Для решения задачи при помощи цикла нам нужно перебрать все символы с конца к началу. Перебирать положительные индексы или отрицательные - неважно. Нужно только правильно указать, на каком индексе цикл должен остановиться. Если ошибиться, то либо слово станет короче, либо программа выдаст ошибку, потому что слово закончится, а цикл попытается продолжить работу.

Если считать с начала слова, первый символ - [0]. А как посчитать индекс последнего символа, используя длину слова?

Можем проверить, правильно ли мы посчитали. Индекс символа в строке можно вывести при помощи метода index(). Метод вызывается на строке (то есть пишется через точку после строки, в которой будем искать символ) и принимает в качестве аргумента то, что мы будем искать.

```python
word = 'сопка'
print(word.index('с'))
print(word.index('о'))
print(word.index('п'))
print(word.index('к'))
print(word.index('а'))

print('длина слова', len(word))
```

Получается, первый символ - `[0]`, а последний - `[len() - 1]`.

Если использовать отрицательные индексы - первый будет `[-1]`, а последний `[-len()]`.

Решим задачу обоими способами.

```python
word = input('Введите любое слово: ')
i = -1
new_word = ''

while i >= -len(word):
    new_word += word[i]
    i -= 1
    
print(new_word)
```

```python
word = input('Введите любое слово: ')
i = len(word) - 1
new_word = ''

while i >= 0:
    new_word += word[i]
    i -= 1
    
print(new_word)
```

Новое слово мы не печатаем сразу, а сначала собираем по букве внутри новой переменной. При этом перед началом цикла мы ввели эту переменную и записали в нее пустую строку. Почему?

* В результате у нашей пустой переменной уже есть тип данных. Она пустая, но уже знает, что она - строка.
* Внутри цикла мы прибавляем символ к какому-то уже существующему значению переменной. Если до цикла переменная существовать не будет, программа не поймет, где брать значение, к которому добавлять символ, и выдаст ошибку.

## Цикл for

Если `while` выполняет цикл, пока условие выполнения истинно, `for` перебирает данные, пока они не закончатся.

```python
fruit = ['яблоко', 'банан', 'слива']

for item in fruit:
    print (item)
```

В переданном циклу списке 3 элемента, поэтому цикл выполнился 3 раза: пока элементы не закончились.

Так же цикл сработает на строке. Элементы строки - символы, поэтому цикл будет перебирать их.

```python
word = 'арбуз'

for letter in word:
    print(letter)
```

Элемент, который цикл перебирает, на каждой *итерации* (то есть на каждом повторении) записывается в переменную, которую мы можем использовать внутри цикла.

**Синтаксис.** За `for` следует переменная, в которую записывается то, что перебирает функция, затем `in` и то, что мы перебираем по элементам. Затем двоеточие и с новой строки с отступом - код цикла.

## Функция range()

`range()` перебирает целые числа в заданном диапазоне. Два аргумента функции задают начало и конец такого перебора. Обратите внимание: начало задается включительно, а конец - не включительно.

```python
for i in range(1, 5):
    print(i)
```

Начало перебора можно не указывать, тогда перебор начнется с 0.

```python
for i in range(5):
    print(i)
```

При необходимости можно задать третий аргумент: шаг перебора. Если третий аргумент не указан, устанавливается его значение по умолчанию, равное 1.

```python
for i in range(1, 10, 2):
    print(i)
```

```python
for i in range(5, 1, -1):
    print(i)
```

`range()`, конечно, можно использовать и для перечисления индексов.

```python
word = 'порог'

for i in range(len(word)):
    print(word[i])
```

`for`, особенно в сочетании с `range()` для многих задач кажется более приятным, поскольку позволяет получить тот же результат меньшим количеством кода. Но и `while` порой оказывается незаменим.

## Срезы

Срезы - способ получить элементы без использования цикла. Выглядит как гибрид индексов и range().

Пишутся в квадратных скобках, первое значение - начало перебора, второе - конец перебора, третье - шаг. Между значениями ставятся двоеточия. Все значения можно пропустить, тогда установятся значения по умолчанию: начало - 0, конец - длина списка или строки, шаг - 1.

```python
word = 'мракоборец'

word[0:4]
```

```python
word[1:10:2]
```

```python
word[:4]
```

```python
word[5:]
```

```python
word[1::2]
```

```python
word[::-1]
```
