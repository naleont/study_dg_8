# Уровень Кроули

Мы с вами уже большие, поэтому можно перестать писать код в IDLE и перейти в более удобную среду разработки, которая станет делать часть работы за вас: например, дописывать команды, исправлять или подсвечивать синтаксис и так далее.

Сред разработки достаточно много, у всех свои особенности. Например, эти уроки я пишу в Jupyter notebook. Этот инструмент удобен, когда нужно исполнять код по частям и видеть, что происходит на разных шагах работы. Аналогичным образом работает Google Collab.

Под наши следующие цели хорошо подойдет VSCode. Это "легкий" редактор, позволяющий работать с проектами. Он уже должен стоять на всех школьных компьютерах. Если программу нужно установить, сделать это можно по [ссылке](https://code.visualstudio.com/).
