# Уровень Халк

Как мы знаем, мысли гениев сходятся. В программировании это тоже актуально. Бывает функционал, который описать при помощи базовых инструментов python возможно, но ооооооочень сложно. Или долго. Или не очень, но все равно не хочется. И точно же кто-то до вас с потребностью именно в этом функционале сталкивался. И этот кто-то уже придумал, как описать этот функционал. Программисты - ленивые люди. И они знают, что другие программисты - такие же. Поэтому они друг с другом делятся. Такие куски кода называют **библиотеками** или **модулями**. Вообще модуль - это частный случай библиотеки, но в контексте практики применения на python это практически одно и то же.

Есть библиотеки, которые программисты считают настолько полезными и часто используемыми, что их сделали предустановленными. Такие библиотеки скачиваются на ваш компьютер, как только вы устанавливаете себе python, и для их использования нужно только сказать программе, что вы будете ими пользоваться, то есть импортировать библиотеку.

Импорт библиотек осуществляется директивой `import`. Например, `import strings` импортирует библиотеку `strings`. Эта библиотека содержит расширенный набор инструментов для работы со строковыми данными. Функции, импортированные из библиотек, пишутся через точку после названия библиотеки. Например, команда `string.ascii_lowercase` выдаст строку `abcdefghijklmnopqrstuvwxyz` - все символы ASCII (American Standard Code for Information Interchange), относящиеся к нижнерегистровым.

Чтобы стандартизировать способы делиться кодом, придумали инструмент pip. С его помощью можно установить библиотеки, которые не предустановлены в python. Такие библиотеки устанавливают через терминал. Сначала нам нужно установить сам инструмент pip, но перед этим проверим, что он еще не был установлен раньше: `pip --version`. Если pip уже установлен, в консоли отобразится его версия. Если нет - сообщение о том, что pip не установлен (что компьютер не знает, что такое pip).

```python
pip --version
```

Если pip не установлен, в терминале вводим команду для его установки: `pip install pip`. Начнется процесс скачивания и установки. После этого можем устанавливать любые библиотеки. У pip есть [сайт](https://pypi.org/project/pip/), на котором описаны все библиотеки и команды для их установки.

## Таблицы и способы работы с ними

Когда в одной жизни ты программируешь, а в другой - работаешь с таблицами, сама по себе напрашивается мысль о том, что нужно обрабатывать таблицы при помощи кода. Ведь функционал редакторов таблиц при всей своей широте в некоторый момент оказывается недостаточным, а писать код на VBA (Visual Basic, встроенный в Excel язык программирования) - это очень неприятное занятие. Но естественный вопрос начинающего программиста - как получить данные не от пользователя, а из таблицы? И как их потом сохранить?

Мы уже научились работать с текстовыми файлами. И есть два формата файлов, которые для компьютера выглядят как текстовые, но позволяют систематизировать информацию как в таблице. Первый из них вам, вероятно, знаком: это формат CSV. Расшифровывается как comma separated values (значения, разделенные запятыми), расширение `.csv`. Каждая строка такой таблицы записана на отдельной строке файла, а между значениями ячеек стоят запятые. Такие файлы мы можем прочитать уже известными нам инструментами, сохранить данные в переменную, разбить при помощи `.split('\n')` на строки, затем на ячейки `.split(',')`, получить списки ячеек внутри списков строк, и работать с этими данными и горя не знать.

Для записи файла, а точнее для подготовки данных к записи, нам пригодится метод, который мы еще не разбирали: `.join()`. В аргументы методу передается список, который нужно объединить в строку. А вызывается этот метод на строке, которую нужно использовать в качестве разделителя между элементами списка.

```python
' '.join(['мама', 'мыла', 'раму'])
```

Напишите программу, которая прочитает этот файл, вычислит попарные произведения чисел внутри каждой строки и запишет их в таблицу третьим столбцом. В результате работы программы должен получиться файл `.csv` с таблицей из трех столбцов, первый из которых совпадает со столбцами исходной таблицы.

Но запятые мы часто используем не для разделения ячеек таблицы, а по их прямому назначению: например, в качестве знака пунктуации в текстовой ячейке или для обозначения десятичных дробей. Когда файл с подобной ячейкой сохраняется, содержимое всей ячейки записывается в кавычки. Благодаря этому при чтении файла не "поедут" столбцы. Но это создает определенную головную боль: нужно понимать, что делать с типами данных, как предусмотреть их единообразие и при этом адекватность, а когда вы записываете ячейки с запятыми в новый файл - не забывать расставлять кавычки в нужных местах.

Поэтому часто оказывается более удобным для работы формат TSV - tab separated values (значения, разделенные табуляциями). Разделитель вместо запятой - табуляция, служебный символ `\t`. Табуляции практически никогда не встречаются внутри ячеек таблиц, поэтому их удобно использовать в качестве разделителей. Для таких файлов нет отдельного расширения, сохраняют их с расширением ‍‍`.txt`, как обычный текстовый файл.
